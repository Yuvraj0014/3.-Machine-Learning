Handelling Imbalance datasets
1- Upsampling method
2- Downsampling method

```python
import numpy as np
import pandas as pd
```

```python
#set the random seed for reproductivity
np.random.seed(123)

#create a DF with two classes
n_samples=1000
class_0_ratio=0.9
n_class_0=int(n_samples*class_0_ratio)
n_class_1=n_samples-n_class_0
```

```python
n_class_0,n_class_1
```

```python
class_0=pd.DataFrame({
    'feature_1':np.random.normal(loc=0,scale=1,size=n_class_0),
    'feature_2':np.random.normal(loc=0,scale=1,size=n_class_0),
    'target':[0]*n_class_0
})

class_1=pd.DataFrame({
    'feature_1':np.random.normal(loc=2,scale=1,size=n_class_1),
    'feature_2':np.random.normal(loc=2,scale=1,size=n_class_1),
    'target':[1]*n_class_1
})
```

```python
pd.concat([class_0,class_1]).reset_index(drop=True)
```

```python
df=pd.concat([class_0,class_1]).reset_index(drop=True)
```

```python
df.head()
```

```python
df['target'].value_counts()
```

##Upsampling

```python
df_minority=df[df['target']==1]
df_majority=df[df['target']==0]
```

```python
df_minority.head()
```

```python
df_majority.head()
```

```python
##performing Upsampling
from sklearn.utils import resample
```

```python
df_minority_upsample=resample(df_minority,
                              replace=True,##sample with replacement
                              n_samples=len(df_majority),# to match the majority class
                              random_state=42
                              )
```

```python
df_minority_upsample.head()
```

```python
df_minority_upsample.shape
```

```python
df_minority_upsample['target'].value_counts()
```

```python
df.upsampled=pd.concat([df_majority,df_minority_upsample])
```

```python
df.upsampled['target'].value_counts()
```

##Downsampling

```python
class_0=pd.DataFrame({
    'feature_1':np.random.normal(loc=0,scale=1,size=n_class_0),
    'feature_2':np.random.normal(loc=0,scale=1,size=n_class_0),
    'target':[0]*n_class_0
})

class_1=pd.DataFrame({
    'feature_1':np.random.normal(loc=2,scale=1,size=n_class_1),
    'feature_2':np.random.normal(loc=2,scale=1,size=n_class_1),
    'target':[1]*n_class_1
})
```

```python
df=pd.concat([class_0,class_1]).reset_index(drop=True)
```

```python
df_minority=df[df['target']==1]
df_majority=df[df['target']==0]
```

```python
df_majority_downsample=resample(df_majority,
                              replace=False,##sample without replacement
                              n_samples=len(df_minority),# to match the minority class
                              random_state=42
                              )
```

```python
df_majority_downsample.shape
```

```python
df=pd.concat([df_minority,df_majority_downsample])
```

```python
df.value_counts()
```
